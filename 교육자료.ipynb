# 색 예시

```python
from matplotlib import colors as mcolors


colors = dict(mcolors.BASE_COLORS, **mcolors.CSS4_COLORS)

# Sort colors by hue, saturation, value and name.
by_hsv = sorted((tuple(mcolors.rgb_to_hsv(mcolors.to_rgba(color)[:3])), name)
                for name, color in colors.items())
sorted_names = [name for hsv, name in by_hsv]

n = len(sorted_names)
ncols = 4
nrows = n // ncols + 1

fig, ax = plt.subplots(figsize=(8, 5))

# Get height and width
X, Y = fig.get_dpi() * fig.get_size_inches()
h = Y / (nrows + 1)
w = X / ncols

for i, name in enumerate(sorted_names):
    col = i % ncols
    row = i // ncols
    y = Y - (row * h) - h

    xi_line = w * (col + 0.05)
    xf_line = w * (col + 0.25)
    xi_text = w * (col + 0.3)

    ax.text(xi_text, y, name, fontsize=(h * 0.8),
            horizontalalignment='left',
            verticalalignment='center')

    ax.hlines(y + h * 0.1, xi_line, xf_line,
              color=colors[name], linewidth=(h * 0.6))

ax.set_xlim(0, X)
ax.set_ylim(0, Y)
ax.set_axis_off()

fig.subplots_adjust(left=0, right=1,
                    top=1, bottom=0,
                    hspace=0, wspace=0)
plt.show()
```

# Tick customizing

https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.tick_params.html

```python
fig = plt.figure()
ax = fig.add_subplot(111)


# add labels

ax.set_xlabel('$x$')
ax.set_ylabel('$y = x^2$')

# reduce the axis limites to let the line touch the borders
ax.set_xlim(0, 10)
ax.set_ylim(0, 100)

# customize the ticks
ax.tick_params(labelleft=False, #y축 글씨 사라짐
               labelcolor='orange', #글자색
               labelsize=12, # 글자크기
               bottom=False, # x축 tick 사라짐
               color='green', #tick 색 바꾸기
               width=4,  
               length=8,
               direction='inout' #tick의 위치
              )

plt.show()
```

subplots_adjust: tune the subplot layout<br/>
left  = the left side of the subplots of the figure<br/>
wspace =the amount of width reserved for space between subplots<br/>
hspace = the amount of height reserved for space between subplots

```python
fig = plt.figure()
ax = fig.add_subplot(111)

x = np.arange(0, 3)
y = data['Embarked'].value_counts()
bars = ax.bar(x, y, color='lightslategrey')

ax.set_frame_on(False) #뒤에 배경 지우기
ax.set_xticks(np.arange(0,3))

ax.tick_params(bottom='off',left='off',labelleft='off')
ax.set_xticklabels(['Southampton','Cherbourg','Queenstown'],{'fontsize':12,'va':'center','ha':'center'})

for bar,value in zip(bars,y):
    ax.text(bar.get_x()+ bar.get_width()/2,
            bar.get_height()-5
            ,value,
            ha='center',
            va='top',
            color='w',
            fontsize=14)

bars[0].set_color('firebrick')

ax.set_title('Most of passengers embarked at Southampton',{'fontsize':18,'fontweight':'bold','va':'baseline',
                                                           'ha':'center'})
```

2.색으로 카테고리를 표현하고 싶은 경우<br/>
cmap에 plt.cm.rainbow 를 넘기는 것이 아니라 cmap=plt.cm.get_cmap('rainbow','color_num')으로 원하는 색만 넘겨줘야<br/>
continuous가 아니라 discrete한 색조합이 나옴.

```python
sample_size = 100
color_num = 3

X = np.random.normal(0, 1, sample_size)
Y = np.random.normal(0, 1, sample_size)
C = np.random.randint(0, color_num, sample_size)

plt.figure(figsize=(12,4))
plt.scatter(X,Y,c=C,s=20,cmap=plt.cm.get_cmap('rainbow',color_num),alpha=0.5)
plt.colorbar(ticks=range(color_num),format='color: %d', label='color')
```

get_cmap(): return the colormap <br />

matplotlib.cm.ScalarMappable(norm=None, cmap=None): <br/>
    norm : Normalize instance <br/>
    cmap : str or Colormap instance



set_bounds():Set the bounds of the rectangle: l,b,w,h

https://matplotlib.org/api/_as_gen/matplotlib.patches.FancyBboxPatch.html?highlight=set_bounds#matplotlib.patches.FancyBboxPatch.set_bounds

```python
fig=plt.figure(figsize=(10,6))
ax=fig.add_subplot(111)

x=data['Age'].dropna()
age_bins=np.arange(0,95,5)
values,bins,bars=ax.hist(x,bins=age_bins)

ax.set_xticks(bins)
ax.set_ylim(values.min(),values.max()+10)

ax.spines['bottom'].set_bounds(bins.min(),bins.max())
ax.spines['right'].set_bounds(values.min(),values.max())

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

cm=plt.cm.get_cmap('viridis')

for i in range(0,len(bins)):
    if i<len(bins)-1:
        survival=data[(data.Age>=bins[i])&(data.Age<bins[i+1])]['Survived'].mean()
    else:
        survival=data[(data.Age>=bins[i])]['Survived'].mean()
    try:
        bars[i].set_color(cm(survival))
    except:
        pass
sm=plt.cm.ScalarMappable(cmap=cm,norm=plt.Normalize(vmin=0,vmax=1))
sm._A=[] #You must first set_array for mappable
plt.colorbar(sm)


```
